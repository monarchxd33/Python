```python
from matplotlib import pyplot
# or (import matplotlib.pyplot)
#importing pyplot(sub module from matplotlib library)
```

```python
age=[10,11,12,13,14,15,16]
height=[120,130,135,150,155,160,170]
```

```python
pyplot.plot(age,height)
#plotting age vs. height
pyplot.xlabel("age")
#giving label to x-axis
pyplot.ylabel("height")
#giving label to y-axis
```

```python
height_us=[125,135,140,145,165,170,180]
pyplot.plot(age,height)
pyplot.plot(age,height_us)
#plotting multiple graph in 1 graph
```

```python
pyplot.plot(age,height,label='india')
pyplot.plot(age,height_us,label='us')
#label parameter is used for identification of curve
pyplot.legend()
#legend is used to identify the parameters like label,linestyle,corners and also used for modification
```

```python
pyplot.plot(age,height,'b',label='india')
pyplot.plot(age,height_us,'k--',label='us')
pyplot.legend()
```

```python
pyplot.plot(age,height,'y',label='india')
pyplot.plot(age,height_us,'r',label='us')
pyplot.legend()
```

```python
pyplot.plot(age,height,'b',linestyle='--',label='india')
pyplot.plot(age,height_us,'k--',linestyle=':',label='us')
pyplot.legend()
#'-','--','_.',':','NOne',solid,dashed,dashdot,dotted
```

```python
pyplot.plot(age,height,'b',linestyle='--',marker='o',label='india')
pyplot.plot(age,height_us,'k--',linestyle=':',marker='s',label='us')
pyplot.legend()

#marker is used for marking all the points with desired shapes
#'s'=square
#'o'=circle
#'^'=triangle
```

```python
pyplot.plot(age,height,'b',linestyle='--',marker='o',label='india')
pyplot.plot(age,height_us,'k--',linestyle=':',marker='s',label='us')
pyplot.grid()
pyplot.legend()
```

```python
#bar graph
pyplot.bar(age,height,label='india')
pyplot.plot(age,height_us,label='us')
pyplot.legend()
```

```python
pyplot.bar(age,height,label='india')
pyplot.bar(age,height_us,label='us')
pyplot.legend()
```

```python
#stackplot
years=[2023,2024,2025,2026,2027]
s1=[1000,1500,2000,4000,5000]
s2=[1100,1500,2200,4000,4800]
s3=[1200,1600,2500,4800,5900]
pyplot.stackplot(years,s1,s2,s3)
```

```python
years=[2023,2024,2025,2026,2027]
s1=[1000,1500,2000,4000,5000]
s2=[1100,1500,2200,4000,4800]
s3=[1200,1600,2500,4800,5900]
pyplot.stackplot(years,s1,s2,s3,labels=['student1','student2','student3'],colors=['black','green','yellow'])
pyplot.xlabel('years')
pyplot.ylabel('income/day')
pyplot.title('income as. years')
pyplot.legend()
```

```python
years=[2023,2024,2025,2026,2027]
s1=[1000,1500,2000,4000,5000]
s2=[1100,1500,2200,4000,4800]
s3=[1200,1600,2500,4800,5900]
pyplot.stackplot(years,s1,s2,s3,labels=['student1','student2','student3'],colors=['black','green','yellow'])
pyplot.xlabel('years')
pyplot.ylabel('income/day')
pyplot.title('income as. years')
pyplot.legend(loc='upper left')

#loc is used for location of the labels and it will take arument like
# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left',
# 'center right', 'lower center', 'upper center', 'center'
```
